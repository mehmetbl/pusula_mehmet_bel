```python
# Chronic diseases by blood type
kronik_hastaliklar = df['Kronik Hastaliklarim'].str.get_dummies(sep=', ')
kan_grubu_hastalik = df[['Kan Grubu']].join(kronik_hastaliklar)
hastalik_sayilari = kan_grubu_hastalik.groupby('Kan Grubu').sum()
plt.figure(figsize=(12, 8))
hastalik_sayilari.plot(kind='bar')
plt.title('Kan Grubuna Göre Kronik Hastalık Sayıları')
plt.xlabel('Kan Grubu')
plt.ylabel('Kronik Hastalık Sayısı')
plt.xticks(rotation=0)
plt.show()
```

```python
# I used SimpleImputer to fill missing data easily. 
# Categorical data is filled with the most frequent value, and numerical data with the mean. 
# This helps me handle missing values quickly and effectively.
```

```python
from sklearn.impute import SimpleImputer
```

```python
kategorik_degiskenler = ['Cinsiyet', 'Il', 'Kan Grubu', 'Alerjilerim', 'Kronik Hastaliklarim',
                       'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari',
                       'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari']

imputer_kat = SimpleImputer(strategy='most_frequent')
df[kategorik_degiskenler] = imputer_kat.fit_transform(df[kategorik_degiskenler])
```

```python
sayisal_degiskenler = ['Kilo', 'Boy']

imputer_say = SimpleImputer(strategy='mean')
df[sayisal_degiskenler] = imputer_say.fit_transform(df[sayisal_degiskenler])
```

```python
# I used One-Hot Encoding to convert categorical data into numbers. 
```

```python
kategorik = ['Cinsiyet', 'Uyruk', 'Il', 'Kan Grubu', 
             'Alerjilerim', 'Kronik Hastaliklarim', 
             'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari',
             'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari']
df_encoded = pd.get_dummies(df, columns=kategorik, drop_first=True)
```

```python
boolean = [col for col in df_encoded.columns if col not in ['Dogum_Tarihi', 'Ilac_Baslangic_Tarihi', 'Ilac_Bitis_Tarihi', 'Yan_Etki_Bildirim_Tarihi']]
```

```python
for col in boolean:
    if df_encoded[col].dtype == 'bool':
        df_encoded[col] = df_encoded[col].astype(int)
```
